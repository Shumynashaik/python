```python
def avg():
    n1=20
    n2=10
    n3=30
    average=n1+n2+n3/3
    print(average)
avg()
```

```python
def add(a,b):
    print(a+b)
def sub(a,b):
    print(a-b)
def mul(a,b):
    print(a*b)
def div(a,b):
    print(a/b)
def cal():
    print('''option1 add,option2 sub,option3 mul,option4 div''')
    option=eval(input("enter the number b/w 1-4:"))
    a=eval(input("enter the number:"))
    b=eval(input("enter the number:"))
    if option==1:
        print("the addition is:")
        add(a,b)
    elif option==2:
        print("the subtraction is:")
        sub(a,b)
    elif option==3:
        print("the multiplication is:")
        mul(a,b)
    else:
        print("the division is:")
        div(a,b)
cal()
```
